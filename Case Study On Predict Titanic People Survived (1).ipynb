```python
x=titanic_data.drop(["survived"],axis=1)
x.head(2)
```

```python
y=titanic_data["survived"].map({0:"Not Survived",1:"Survived"})
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)
print(f"x_train is {x_train.shape}")
print(f"x_test is {x_test.shape}")
print(f"y_train is {y_train.shape}")
print(f"y_test is {y_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model_logr=LogisticRegression()
model_logr
```

```python
model_logr.fit(x_train,y_train)
```

```python
lg_y_pred=model_logr.predict(x_test)
lg_y_pred
```

```python
tree = DecisionTreeClassifier()
tree
```

```python
tree.fit(x_train,y_train)
```

```python
dt_y_pred = tree.predict(x_test)
dt_y_pred
```

```python
rf = RandomForestClassifier()
rf
```

```python
rf.fit(x_train,y_train)
```

```python
rf_y_pred=rf.predict(x_test)
rf_y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f"Logistic Regression {accuracy_score(y_test, lg_y_pred)}")
print(f"Decision Tree Classifier {accuracy_score(y_test, dt_y_pred)}")
print(f"Random Forest Classifier  {accuracy_score(y_test, rf_y_pred)}")
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, lg_y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print("---------------------Logistic Regression----------------------------")
print(classification_report(y_test,lg_y_pred))
```

```python
print("---------------------Decision Tree Classifier----------------------------")
print(classification_report(y_test,dt_y_pred))
```

```python
print("---------------------Random Forest Classifier----------------------------")
print(classification_report(y_test,rf_y_pred))
```
